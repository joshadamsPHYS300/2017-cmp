Try to do the intergral many times (100) for different $N$ values (100, 1000, 10000, 100000), collect the data and calculate its standard varaince. (use 13.4 integral, use np.std and square it for variance)

```python
#Homework, part 1.

import matplotlib.pyplot as plt
import numpy as np
from random import random

fig1 = plt.figure()
ax1 = fig1.add_subplot(111, aspect='equal')

x1 = np.linspace(0,2,10000)
y1 = np.sqrt(1-(x1-1)**2) #small circle
ax1.plot(x1,y1,'k')
y2 = 2-np.sqrt(4-x1**2) #big circle
ax1.plot(x1,y2,'k')
ax1.set_xlim(0,1.6)
ax1.set_ylim(0,1.1)
ax1.fill_between(x1, y1, y2, where=y1>y2, facecolor='green')
plt.show()

def fb(x):
    return 2-np.sqrt(4-x**2)
def ft(x):
     return np.sqrt(1-(x-1)**2)
def MC(N):
    count = 0
    for i in range (N):
            x=2*random()
            y=2*random()
            if fb(x)<y and y<ft(x):
                count += 1
    I= 4*count/N
    return I

print(MC(100),MC(1000),MC(10000),MC(100000))
```

```python
#Homework, part 2

# generate plot of the function
import matplotlib.pyplot as plt
import numpy as np

def f(x):
    return np.sin(1/x/(2-x))**2

# plot the f(x)
x = np.linspace(0.001,2,100)
plt.plot(x, f(x))
# plot the boundary
plt.plot([0,0,2,2,0],[0,1,1,0,0])
plt.show()

# MC integral
import matplotlib.pyplot as plt
from math import sin
from random import random

def f(x):
    return sin(1/x/(2-x))**2
def MC(N):
    count = 0
    for i in range(N):
        x = 2*random()
        y = random()
        if y<f(x):
            count += 1
    I = 2*count/N
    return I

#define lists to store data sets:
n_100 = []
n_1000 = []
n_10000 = []
n_100000 = []

for z in range(100):
    n_100.append(MC(100))
    n_1000.append(MC(1000))
    n_10000.append(MC(10000))
    n_100000.append(MC(100000))

n_100 = np.array(n_100)
print('For N = 100:' ', Integral =', min(n_100), ', Standard Deviation =',np.std(n_100), ', Variance =',np.std(n_100)**2)
print('For N = 1000:' ', Integral =', min(n_1000), ', Standard Deviation =',np.std(n_1000), ', Variance =',np.std(n_1000)**2)
print('For N = 10000:' ', Integral =', min(n_10000), ', Standard Deviation =',np.std(n_10000), ', Variance =',np.std(n_10000)**2)
print('For N = 100000:' ', Integral =', min(n_100000), ', Standard Deviation =',np.std(n_100000), ', Variance =',np.std(n_100000)**2)
```
