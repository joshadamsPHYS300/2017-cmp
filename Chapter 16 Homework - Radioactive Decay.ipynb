- 1, rewrite the radioactive decay, to generate 1000 random numbers following the non-uniformed distribution.

```python
import numpy as np
import matplotlib.pyplot as plt

N = 1000
tau = 3.053*60
mu = np.log(2)/tau

x = np.random.random(N)

t_dec = (-1/mu)*(np.log(1-x))
t_dec = np.sort(t_dec)
dec = np.arange(1,N+1)

sur = -dec + N

plt.plot(t_dec,sur,label='1000 Random Numbers')
plt.xlabel("Time")
plt.ylabel("Number of atoms")
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

NTl = 1000
NPb = 0
tau = 3.053*60
dt = 1 
p = 1 - 2**(-dt/tau)
tmax = 1000

t_points = np.arange(0.0, tmax, dt)
Tl = []
Pb = []

for t in t_points:
    Tl.append(NTl)
    Pb.append(NPb)
    
    decay = 0
    for i in range(NTl):
        if np.random.random()<p:
            decay += 1
    NTl -= decay
    NPb += decay

plt.plot(t_points, Tl, label='Tl')
plt.plot(t_points, Pb, label='Pb')
plt.plot(t_dec,sur, label='Revised')
plt.xlabel("Time")
plt.ylabel("Number of atoms")
plt.legend()
plt.show()
```
