## 17.5 Homework

Write a code to solve the minimum of 2D function with Monte Carlo method.


```python
import numpy as np
import matplotlib.pyplot as plt

x_min, x_max = -2, 2
y_min, y_max = -2, 2

def f(x,y):
    return 1/2*x**2 + 1/4*y**2

nx = np.linspace(x_min,x_max,100)
ny = np.linspace(y_min,y_max,100)

x,y = np.meshgrid(nx,ny)

z = f(x, y)
fig = plt.figure()
ax = fig.gca(projection='3d')
ax.plot_surface(x,y,z, cmap='summer', alpha=0.8)
cset = ax.contour(x,y,z, zdir='z', offset=0)
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('f')

plt.show()

def g(x):
    return 1/2*x**2 
def h(y):
    return 1/4*y**2
def MC(N=1000): #Monte Carlo
    x1 = 2*np.random.random(N)
    y1 = np.random.random(N)
    f1 = np.delete(x1,x1!=g) #delete returns a new array with sub-arrays along an axis deleted.
    f2 = np.delete(y1,y1!=h)
    minimum_x = np.amin(f1) #amin returns the minimum along an axis.
    minimum_y = np.amin(f2)
    min = f(minimum_x,minimum_y)
    print('Minimum =',min) 
MC()
```
